```python
title_list=[]
company_list=[]
location_list=[]
date_list=[]
link_list=[]


for job_element in job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="subtitle is-6 company")
    location_element = job_element.find("p", class_="location")
    date_element=job_element.find("time")
    link_element=job_element.find('a', href=True, string="Apply")
    link= link_element['href']
    
    
    link_list.append(link)
    title_list.append(title_element.text)
    company_list.append(company_element.text)
    location_list.append(location_element.text)
    date_list.append(date_element.text)
```

```python
def get_newslist( ):
    newslist2=[]
    title_list2=[]
    date_list2=[]
    
    page_source = driver.page_source
    time.sleep(0.5)
    soup = BeautifulSoup(page_source)
    
    linklist = soup.find_all('h1', class_='or-newsroom-article-headline')
    ddlist=soup.find_all('div',class_='or-newsroom-article-date')
    
    for link in linklist:
        linkcontent = link.find('a', href=True)
        
        title=linkcontent.text
        title_list2.append(title)
        
        linkcontent = linkcontent['href']
        newslist2.append(linkcontent)
    K='https://www.oregon.gov/newsroom/Pages/'
    newslist2= [K+str(x) for x in newslist2]
        
    for dd in ddlist:
        date=dd.text
        date_list2.append(date)
        
    return newslist2, title_list2, date_list2
```
